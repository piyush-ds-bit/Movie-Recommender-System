```python
import pandas as pd
import ast 
```

```python
movies=pd.read_csv('~/DATA Sets/tmdb_5000_movies.csv')
credit=pd.read_csv('~/DATA Sets/tmdb_5000_credits.csv')
movies=movies.merge(credit,on='title')
```

```python
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.dropna(inplace=True)
```

```python
movies.head()
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L

movies['genres']=movies['genres'].apply(convert)
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def convert4(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=="Director":
            L.append(i['name'])
            break
    return L
movies['crew']=movies['crew'].apply(convert4)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['genres']+movies['keywords']+movies['cast']+movies['crew']+movies['overview']
```

```python
new_data_frame=movies[['movie_id','title','tags']]
```

```python
new_data_frame
```

```python
new_data_frame['tags']=new_data_frame['tags'].apply(lambda x:" ".join(x))
new_data_frame
```

## Data Preprocessing is now completed,now we will move onto making the model.....

```python
from sklearn.feature_extraction.text import CountVectorizer
narayan=CountVectorizer(max_features=5000,stop_words='english')
sri_narayan=narayan.fit_transform(new_data_frame['tags']).toarray()
```

```python
from nltk.stem.porter import PorterStemmer
jai_sri_narayan=PorterStemmer()
```

```python
def stemming(text):
    y=[]
    for i in text.split():
        y.append(jai_sri_narayan.stem(i))
    return " ".join(y)
new_data_frame['tags']=new_data_frame['tags'].apply(stemming)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(sri_narayan)
```

```python
def recommend(movie):
    movie_index=new_data_frame[new_data_frame['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_data_frame.iloc[i[0]].title)
```

## Lets us check,how the model is recommending

```python
hii=recommend("The Dark Knight Rises")
hii
```
